```python
df = df.drop_duplicates()
df = df.dropna()
```

```python
import re

def convert_price(price_str):
    """
    Extracts the numeric part from a price string and converts it to a full numeric value.
    For example, "PKR 68.5 lacs" becomes 6850000.0 (if 1 lac = 100,000).
    """
    match = re.search(r'([\d\.]+)', price_str)
    if match:
        value = float(match.group(1))
        if 'lacs' or 'crore' in price_str.lower():
            value *= 100000  # 1 lac = 100,000
        return value
    return None
```

```python
# Converting the 'Price' column using the custom function
#df['Price_numeric'] = df['Price'].apply(convert_price)
df.loc[:, 'Price_numeric'] = df['Price'].apply(convert_price)
df
```

```python
df = df.drop('Price', axis=1)
```
